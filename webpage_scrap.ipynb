```python
missing_links = []
senator_history_lst = []

for full_file_link_path in file_lst_for_data:
  #get xml data for each xml file link
  r = requests.get(full_file_link_path)
  print(r.status_code)
  #print(r.text)

  #make sure we got a file....
  if(r.status_code == 200):
    #get the data....
    vote_details = xmltodict.parse(r.text)
    congress_id_num = vote_details['roll_call_vote']['congress']
    congress_session_num = vote_details['roll_call_vote']['session']
    vote_call_num = vote_details['roll_call_vote']['vote_number']
    vote_call_issue = vote_details['roll_call_vote']['vote_title']

    #setup loop for what each senator voted...
    senator_member_lst = vote_details['roll_call_vote']['members']['member']
    print('Currently working on: ', vote_call_num)


    for each_sen_mem in senator_member_lst:
      #print(each_sen_mem['last_name'])
      new_row_dict = {}
      new_row_dict['congress'] = congress_id_num
      new_row_dict['session'] = congress_session_num
      new_row_dict['vote_number'] = vote_call_num
      new_row_dict['vote_title'] = vote_call_issue
      new_row_dict['last_name'] = each_sen_mem['last_name']
      new_row_dict['party'] = each_sen_mem['party']
      new_row_dict['vote_cast'] = each_sen_mem['vote_cast']

      #add to list for later examation
      #after building list, will add to dataframe.
      senator_history_lst.append(new_row_dict)

  else:
    #something went wrong...
    #keep track of this problem
    missing_links.append(full_file_link_path)

  #before moving to the next record, pause for some time....
  time.sleep(1)
```
