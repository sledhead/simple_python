```python
df = pd.read_excel('Excel_Project_Dataset.xlsx')

df.head(10)
```

```python
df['Children'].agg(['mean','min','max'])
```

```python
df['Age'].agg(['mean','min','max'])
```

```python
df['Income'].agg(['mean','min','max'])
```

```python
df.groupby(['Occupation']).agg({'Income':['mean','min','max']}).round(2)
```

```python
agg_func_describe = {'Income': ['describe']}
df.groupby(['Occupation']).agg(agg_func_describe).round(2)
```

```python
agg_func_describe = {'Age': ['describe']}
df.groupby(['Occupation']).agg(agg_func_describe).round(2)
```

```python
agg_func_count = {'Age': ['count', 'nunique', 'size']}
df.groupby(['Occupation']).agg(agg_func_count)
```
